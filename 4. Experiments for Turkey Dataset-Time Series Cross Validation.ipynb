**Load Data**

```python
import pandas as pd
df = pd.read_csv('/root/CEEMDAN-EWT-LSTM/dataset/T1.csv')
new_data = df.dropna(axis = 0, how ='any')

# Xử lý dữ liệu Turkey: Tạo cột 'month' từ Date/Time
# Format: "01 01 2018 00:00" -> parse thành datetime
new_data['Date'] = pd.to_datetime(new_data['Date/Time'], format='%d %m %Y %H:%M', errors='coerce')
new_data['Year'] = new_data['Date'].dt.year
new_data['month'] = new_data['Date'].dt.month  # Tạo cột 'month' (chữ thường) để tương thích với myfunctions.py
new_data = new_data[['month', 'Year', 'Date', 'LV ActivePower (kW)']]  # Chọn các cột cần thiết

new_data.describe()
cap=max(new_data['LV ActivePower (kW)'])
```
